```python
non_continous_data = ["Country","Region"]


continuousData = dataset[[col for col in dataset.columns if col not in non_continous_data]]
continuousData
```

```python
import seaborn as sns # Used for plotting

corr_ignore = ["Economy_status_Developed"]

correlation_data = continuousData[[a for a in continuousData.columns if a not in corr_ignore]]

corr = correlation_data.corr()

print(corr["Population_mln"].round(2))


mask = np.triu(corr)
mask[np.diag_indices_from(mask)] = False

plt.rcParams['figure.figsize'] = (11, 11)

sns.heatmap(
    corr, 
    annot = True,
    xticklabels = corr.columns.values,
    yticklabels = corr.columns.values,
    mask = mask);

plt.title('Correlation between Attributes', fontsize = 13);
plt.xticks(fontsize = 12);
plt.yticks(fontsize = 12);
```
